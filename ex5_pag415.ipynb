```
# Capítulo 9, seção 9-4, exercício 5
# Esboço do gráfico de f(x) e sua série de Fourier

from matplotlib import pyplot as plt
from math import pi, sin, cos

h = 0.1 # passo entre os pontos
# Vetores que armazenam os pontos 'x' e seus respectivos valores 'f(x)' de acordo com o intervalo dado
Xf1, Yf1 = [], [] # -pi até -1/2
Xf2, Yf2 = [], [] # -1/2 até 1/2
Xf3, Yf3 = [], [] # 1/2 até pi
x = -pi
# Salvando pontos de f(x)
while x <= pi:
  if -pi <= x <= -1/2:
    Xf1.append(x)
    Yf1.append(-abs(x-1/2))
    x += h
  elif -1/2 <= x <= 1/2:
    Xf2.append(x)
    Yf2.append(0)
    x += h
  elif 1/2 <= x <= pi:
    Xf3.append(x)
    Yf3.append(x-1/2)
    x += h
# Salvando pontos da série de Fourier de f(x)
x_fourier, y_fourier = [], []
x = -pi
while x <= pi:
  y = 1/4*pi - 1/2
  for k in range(1,9):
    y += (sin(k/2)/k*pi)*cos(k*x) + (pow(-1,k+1)*2/k + cos(k/2)/k*pi - 2*sin(k/2)/pow(k,2)*pi)*sin(k*x)
  x_fourier.append(x)
  y_fourier.append(y)
  x += h
# Plotando gráficos
plt.plot(Xf1, Yf1, color='blue', label='f(x)')
plt.plot(Xf2, Yf2, color='blue')
plt.plot(Xf3, Yf3, color='blue')
plt.plot(x_fourier, y_fourier, color='orange', label="Série de Fourier")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc="upper left")
plt.grid("True")
plt.show()
```
