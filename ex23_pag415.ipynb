```
# Exercício 23, letras a b c, capítulo 9, seção 9-4, página 415
# Plotando gráficos das funções e suas séries de Fourier

from matplotlib import pyplot as plt
from math import pi, sin, cos

# Vetores de pontos x,f(x) da letra 'a'
X1f1, Y1f1 = [], []
X2f1, Y2f1 = [], []
x_fourier_f1, y_fourier_f1 = [], []
# Vetores de pontos x,f(x) da letra 'b'
X1f2, Y1f2 = [], []
X2f2, Y2f2 = [], []
x_fourier_f2, y_fourier_f2 = [], []
# Vetores de pontos x,f(x) da letra 'c'
X1f3, Y1f3 = [], []
X2f3, Y2f3 = [], []
x_fourier_f3, y_fourier_f3 = [], []
# Calculando e armazenando pontos (x,f(x))
h = 0.1 # distância entre os pontos
x = -pi + 0.1
while x < pi:
  if - pi < x < 0:
    X1f1.append(x), X1f2.append(x), X1f3.append(x)
    Y1f1.append(1/2), Y1f2.append(x+pi), Y1f3.append(-x)
    x += h
  elif 0 < x < pi:
    X2f1.append(x), X2f2.append(x), X2f3.append(x)
    Y2f1.append(-1/2), Y2f2.append(x), Y2f3.append(-x+2*pi)
    x += h
  else:
    x = 0.1
# Calculando e armazenando pontos das séries de Fourier
x = -pi
h = 0.01
while x < pi:
  y1 = 0
  y2 = pi/2
  y3 = pi
  for k in range(1,15):
    y1 += (-2/pi)*((sin((2*k-1)*x)/2*k-1))
    y2 += (-1/k)*(sin(2*k*x))
    y3 += (2/k)*(sin(k*x))
  x_fourier_f1.append(x), y_fourier_f1.append(y1)
  x_fourier_f2.append(x), y_fourier_f2.append(y2)
  x_fourier_f3.append(x), y_fourier_f3.append(y3)
  x += h
# Plotando gráficos
figura, graficos = plt.subplots(3)
figura.suptitle("Laranja - Série de Fourier\n Azul - f(x)")

graficos[0].plot(X1f1, Y1f1, color='blue')
graficos[0].plot(X2f1, Y2f1, color='blue')
graficos[0].plot(x_fourier_f1, y_fourier_f1, color='orange')
graficos[0].grid("True")

graficos[1].plot(X1f2, Y1f2, color='blue')
graficos[1].plot(X2f2, Y2f2, color='blue')
graficos[1].plot(x_fourier_f2, y_fourier_f2, color='orange')
graficos[1].grid("True")

graficos[2].plot(X1f3, Y1f3, color='blue')
graficos[2].plot(X2f3, Y2f3, color='blue')
graficos[2].plot(x_fourier_f3, y_fourier_f1, color='orange')
graficos[2].grid("True")

plt.show()
```
